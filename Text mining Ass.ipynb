```python
import numpy as np
import pandas as pd
import string 
import spacy

from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
from wordcloud import WordCloud
%matplotlib inline
```

```python
import pandas 
musk= pd.read_csv("C:\\Users\\PRATIK\\Downloads\\Elon_musk.csv",error_bad_lines=False)
```

```python
musk
```

```python
musk = [Text.strip() for Text in musk.Text]
musk = [Text for Text in musk if Text]
musk[0:10]
```

```python
musk
```

```python

text = ' ' .join(musk)
text
```

```python

no_punc_text = text.translate(str.maketrans('','',string.punctuation))
no_punc_text
```

```python
import nltk
nltk.download('punkt')
```

```python

from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
print(text_tokens[0:50])
text_tokens
```

```python
len(text_tokens)
```

```python

from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
my_stop_words = stopwords.words("english")
my_stop_words.append('the')
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:40])
```

```python
my_stop_words
```

```python
len(no_stop_tokens)
```

```python

lower_words = [x.lower() for x in no_stop_tokens]
print(lower_words[0:25])
```

```python
len(lower_words)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
stemmed_tokens = [ps.stem(word) for word in lower_words]
print(stemmed_tokens[0:40])
```

```python
doc =nlp(' '.join(stemmed_tokens))
print(doc[0:40])
```

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas[0:25])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(lemmas)
```

```python
print(vectorizer.vocabulary_)
```

```python
print(vectorizer.get_feature_names()[50:100])
print(x.toarray()[50:100])
```

```python
print(x.toarray().shape)
```

```python
x.toarray()
```

```python
vectorizer_ngram_range = CountVectorizer(analyzer ='word', ngram_range=(1,3),max_features =100)
bow_matrix_ngram = vectorizer_ngram_range.fit_transform(musk)
```

```python
#import packages
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
```

```python
#Define a function to plot word cloud
def plot_cloud(wordcloud):
    #set figure zise
    plt.figure(figsize=(40,30))
    #Display image
    plt.imshow(wordcloud)
    #No axis details
    plt.axis("off");
```

```python
#Generate wordcloud
stopwords = STOPWORDS
stopwords.add('will')
stopwords.add('apple')
stopwords.add('Amazon')
stopwords.add('laptop')

wordcloud = WordCloud(width=3000, height=2000,background_color='black',max_words=100,
                     colormap='Set2',stopwords=stopwords).generate(text)

#plot
plot_cloud(wordcloud)
```
